```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import games
import qlearn
```

```python
game = games.matching_pennies()

reward = 0.05
gamma = 0.0
alpha_param = [100, 1/10000]
delta_param = [20000, 1, 6]
iter_max = 100000
explore_rate = 0.05

run = qlearn.PHC(game, reward, gamma, alpha_param, delta_param,
                 iter_max, explore_rate, True)

print('Initial PDF: \n', run[0], '\n')
print('Converged PDF: \n', run[1])
```

```python
plt.plot(run[2][0, 0, :])
```
